## Printing
Python lets you print more complicated things by combining them into one big string using "string concatenation"

```python
print('Our float value is ' + str(f) + '. Our int value is ' + str(i)) 
```

Alternatively, you can use the much nicer syntax of "string interpolation"

```python
print(f'Our float value is {f}. Our int value is {i}.')
```

## Conditional statements in python

```python
if i == 1 and f > 4:
    print('A: The value of i is 1 and f is greater than 4.')
elif i > 4 or f > 4:
    print('B: i or f are both greater than 4.')
else:
    print('C: both i and f are less than or equal to 4')
```

## Conditional loops

```python
print(l)
```

```python
for e in l:
    print(e)
```

```python
for i in range(len(l)):
    print(i, l[i])
```

Note that in Python, we don't use {} or other markers to indicate the part of the loop that gets iterated.  Instead, we just indent and align each of the iterated statements with spaces or tabs. (You can use as many as you want, as long as the lines are aligned.)

```python
counter = 6
while counter < 10:
    print(counter)
    counter += 1
```

## Creating functions in Python

Again, we don't use {}, but just indent the lines that are part of the function.

```python
def add2(x):
    y = x + 2
    return y
```

```python
i = 5
```

```python
add2(i)
```

### Lambda Function

```python
square = lambda x: x*x
```

```python
print(square(5))
```
